```python
# Function to read the first n lines of the JSONL file and create a DataFrame
def create_dataframe(file_path, num_lines=2000):
    data = []
    with open(file_path, 'r') as file:
        for i, line in enumerate(file):
            if i >= num_lines:
                break
            data.append(json.loads(line))
    
    # Create DataFrame
    df = pd.DataFrame(data)
    return df

# Create DataFrame with the first 2000 rows
df = create_dataframe(file_path, num_lines=2000)

# Display the first few rows of the DataFrame
df.head()
```

```python
# Function to normalize a single entry
def normalize_entry(entry):
    normalized_entry = {}
    
    # Flatten addresses
    if 'addresses' in entry and entry['addresses']:
        address = entry['addresses'][0]
        normalized_entry['address'] = address.get('address', '')
        normalized_entry['country'] = address.get('country', '')
        normalized_entry['type'] = address.get('type', '')

    # Flatten identifiers
    if 'identifiers' in entry and entry['identifiers']:
        identifiers = entry['identifiers']
        normalized_entry['id'] = identifiers[0].get('id', '')
        normalized_entry['schemeName'] = identifiers[0].get('schemeName', '')
        if len(identifiers) > 1:
            normalized_entry['id_ent'] = identifiers[1].get('id', '')
            normalized_entry['schemeName_ent'] = identifiers[1].get('schemeName', '')
            normalized_entry['uri_ent'] = identifiers[1].get('uri', '')

    # Flatten names
    if 'names' in entry and entry['names']:
        name = entry['names'][0]
        normalized_entry['fullName'] = name.get('fullName', '')
        normalized_entry['name_type'] = name.get('type', '')

    # Flatten nationalities
    if 'nationalities' in entry and entry['nationalities']:
        nationality = entry['nationalities'][0]
        normalized_entry['nat_code'] = nationality.get('code', '')
        normalized_entry['nat_name'] = nationality.get('name', '')

    # Add personType
    normalized_entry['personType'] = entry.get('personType', '')

    # Flatten publicationDetails
    if 'publicationDetails' in entry:
        publication = entry['publicationDetails']
        normalized_entry['bodsVersion'] = publication.get('bodsVersion', '')
        normalized_entry['license'] = publication.get('license', '')
        normalized_entry['publicationDate'] = publication.get('publicationDate', '')
        if 'publisher' in publication:
            publisher = publication['publisher']
            normalized_entry['pblshr_name'] = publisher.get('name', '')
            normalized_entry['pblshr_url'] = publisher.get('url', '')

    # Flatten source
    if 'source' in entry:
        source = entry['source']
        normalized_entry['src_assertedBy'] = source.get('assertedBy', '')
        normalized_entry['src_description'] = source.get('description', '')
        normalized_entry['src_retrievedAt'] = source.get('retrievedAt', '')
        normalized_entry['src_type'] = source.get('type', '')
        normalized_entry['src_url'] = source.get('url', '')

    # Add statementDate, statementID, and statementType
    normalized_entry['statementDate'] = entry.get('statementDate', '')
    normalized_entry['statementID'] = entry.get('statementID', '')
    normalized_entry['statementType'] = entry.get('statementType', '')

    # Flatten incorporatedInJurisdiction
    if 'incorporatedInJurisdiction' in entry:
        jurisdiction = entry['incorporatedInJurisdiction']
        normalized_entry['Jurisd_code'] = jurisdiction.get('code', '')
        normalized_entry['Jurisd_country'] = jurisdiction.get('name', '')

    return normalized_entry
```

```python
# Read and normalize each entry from the raw JSON file, then write to a new JSON file
with open(file_path, 'r') as infile, open(output_notfiltered_json_path, 'w') as outfile:
    for line in infile:
        entry = json.loads(line)
        normalized_entry = normalize_entry(entry)
        outfile.write(json.dumps(normalized_entry) + '\n')

print(f"Normalized data saved to {output_notfiltered_json_path}")
```

```python
# Path to the normalized JSON file
normalized_json_path = '/Users/pc/Documents/Study/Phyton/Projects/NotFiltered_REG.json'

# List of Ukrainian cities and regions with common alternative spellings
ukrainian_locations = [
    'Kyiv', 'Kiev', 'Kyjev', 'Kharkiv', 'Kharkov', 'Charkiv', 'Odessa', 'Odesa', 'Dnipro', 'Dnepr', 'Donetsk',
    'Zaporizhia', 'Zaporozhye', 'Lviv', 'Lvov', "L'viv", 'Kryvyi Rih', 'Krivoy Rog', 'Mykolaiv', 'Nikolaev',
    'Mariupol', 'Luhansk', 'Lugansk', 'Vinnytsia', 'Vinnitsa', 'Simferopol', 'Kherson',
    'Poltava', 'Chernihiv', 'Chernigov', 'Cherkasy', 'Cherkassy', 'Sumy', 'Zhytomyr', 'Zhitomir',
    'Khmelnytskyi', 'Khmelnytsky', 'Chernivtsi', 'Chernovtsy', 'Rivne', 'Rovno', 'Ivano-Frankivsk',
    'Ternopil', 'Kropyvnytskyi', 'Kropyvnytsky', 'Lutsk', 'Uzhhorod', 'Uzhgorod'
]

# Function to check if an address is in Ukraine
def is_address_in_ukraine(address):
    for location in ukrainian_locations:
        if location in address:
            return True
    return False

# Function to filter entries with Ukrainian roots
def filter_ukrainian_roots(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            entry = json.loads(line)
            
            # Check for Ukrainian roots in various fields
            if (
                (entry.get('nat_code') == 'UA') or
                (entry.get('country') == 'UA') or
                (entry.get('Jurisd_code') == 'UA') or
                ('UA' in entry.get('schemeName', '') or 'UA' in entry.get('schemeName_ent', '')) or
                (is_address_in_ukraine(entry.get('address', '')))
            ):
                data.append(entry)
    return data

# Filter data with Ukrainian roots from normalized data
filtered_data = filter_ukrainian_roots(normalized_json_path)

# Create DataFrame from filtered data
df_filtered = pd.DataFrame(filtered_data)

# Display the first few rows of the filtered DataFrame
pd.set_option('display.max_colwidth', None)
print(df_filtered.head())
```
